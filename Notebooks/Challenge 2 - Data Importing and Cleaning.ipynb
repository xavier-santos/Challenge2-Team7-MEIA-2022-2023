## Importing data

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../Data/dataset.csv', na_values=['#EMPTY?'], delimiter=";")
```

```python
print(df.head(5))
```

```python
print("Number of unique engine IDs:")
print(df["id"].unique().size)
```

## Calculating TTF (Time to Failure)
In this case number of cycles remaining

```python
df["ttf"] = df.groupby("id")["cycle"].transform("max")- df["cycle"]
```

```python
print(df.head(5))
```

## Remove ID Column

```python
df = df.drop("id", 1)
print(df)
```

## Shuffle data

```python
df_shuffled = df.sample(frac=1)
print(df_shuffled.head(5))
```
